# Group By

The groupby() Function In Pandas Is Used To Group Data Based On One Or More Columns In DataFrame

```python
df.groupby(['Size'], as_index= False)['Qty'].sum().sort_values(by= 'Qty', ascending= False)
```

```python
size_qty = df.groupby(['Size'], as_index= False)['Qty'].sum().sort_values(by= 'Qty', ascending= False)
sns.barplot(x= 'Size', y= 'Qty', data=size_qty )
```

```python
sns.countplot(data=df, x='Courier Status', hue='Status')
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10, 5))
plt.hist(column_data, bins=15, edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

###### Note: From The Graph Above You Can See Most Poeple Bought T-shirts

```python
# Checking B2B Data By Using Pie Chart

B2B_check = df['B2B'].value_counts()

# Plot The Pie Chart
plt.pie(B2B_check, labels=B2B_check.index, autopct='%1.1f%%')

#plt.axis('equal')

plt.show()
```

Business-to-business (B2B) is a transaction or business conducted between one business and another, such as a wholesaler and retailer.

###### Note: From Above Chart We Can See That Maximum i.e. 99.2% Of Buyers Are Retailers And 0.8% Are B2B Buyers.

```python
# PrePare Data For Scatter Plot

x_data = df['Category']
y_data = df['Size']


# Plot The Scatter Plot
plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
# Plot Count Of Cities By State

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('Ship State')
plt.ylabel('Count')
plt.title('Distribution Of State')
plt.xticks(rotation=90)
plt.show()
```

```python
# Top 10 States
top_10_states = df['ship-state'].value_counts().head(10)

# Plot Count Of Cities By State
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['ship-state'].isin(top_10_states.index)], x='ship-state')
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.xticks(rotation=45)
plt.show()
```

###### Note: From Above Graph You Can See That Most Of The Buyes Are From Maharashtra

##### Conclusion

The Data Analysis Reveals That The Business Has a Significant Customer Base In Maharashtra State, Mainly Serves Retailers, Fulfills Orders Through Amazon, Experiences High Demand For T-shirts, And Sees M-Size As The Preferred Choice Among Buyers.
